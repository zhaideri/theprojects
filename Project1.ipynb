```python
import random
import csv
from datetime import datetime

# Define the Product, Order, Customer, Store, and Corporation classes
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Order:
    def __init__(self, order_id):
        self.order_id = order_id
        self.products = []
        self.date = datetime.now().date()
        self.time = datetime.now().time()

    def add_product(self, product):
        self.products.append(product)

class Customer:
    def __init__(self, customer_id):
        self.customer_id = customer_id
        self.orders = []

    def create_order(self, order):
        self.orders.append(order)

class Store:
    def __init__(self, store_id):
        self.store_id = store_id
        self.customers = []

    def add_customer(self, customer):
        self.customers.append(customer)

class Corporation:
    def __init__(self, name):
        self.name = name
        self.stores = []

    def add_store(self, store):
        self.stores.append(store)

    def write_sales_data_to_csv(self, filename):
        with open(filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            # Updated headers to include Date and Time
            writer.writerow(["Date", "Time", "Store ID", "Customer ID", "Order ID", "Product Name", "Product Price"])

            for store in self.stores:
                for customer in store.customers:
                    for order in customer.orders:
                        for product in order.products:
                            writer.writerow([order.date, order.time, store.store_id, customer.customer_id, order.order_id, product.name, product.price])

# Create an instance of Corporation with the name "Technology Inc."
corp = Corporation(name="Technology Inc.")

# Main script to generate data
if __name__ == "__main__":
    # Define the number of stores and customers per store
    num_stores = 5
    customers_per_store = 100
    orders_per_customer = 3
    products_per_order = 2

    # Define a list of sample products to choose from
    sample_products = [
        Product(name="Flatscreen TV", price=1000.00),
        Product(name="Hologram", price=9000.00),
        Product(name="VR goggles", price=200.00),
        Product(name="Headphones", price=350.00),
        Product(name="Universal Charging Cable", price=20.50),
        Product(name="Silicon Ornament", price=20.00)
    ]

    # Generate stores and customers
    for store_id in range(1, num_stores + 1):
        store = Store(store_id=store_id)
        corp.add_store(store)

        # Generate customers for each store
        for customer_id in range(store_id * 100, store_id * 100 + customers_per_store):
            customer = Customer(customer_id=customer_id)
            store.add_customer(customer)

            # Generate orders for each customer
            for _ in range(orders_per_customer):
                # Ensure unique order IDs by using a combination of store and customer info
                order_id = f"{store_id}-{customer_id}-{random.randint(10, 9999)}"
                order = Order(order_id=order_id)
                customer.create_order(order)

                # Add random products to the order
                for _ in range(products_per_order):
                    product = random.choice(sample_products)
                    order.add_product(product)

    # Write the generated sales data to CSV
    corp.write_sales_data_to_csv("zain_data.csv")
```
